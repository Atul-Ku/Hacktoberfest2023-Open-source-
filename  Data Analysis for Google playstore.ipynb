**DATA ANALYSIS FOR AUTOMATIC CATEGORY QUESTIONING.**

```python
df=data['Category']
```

```python
for i in df.unique():
    ct=0
    for j in df:
        if(j==i):
            ct+=1
    print("There are",ct,"number of items in",i,"category.")
```

```python
# For storing it in Dictionary.

category={}
for i in df.unique():
    ct=0
    for j in df:
        if(j==i):
            ct+=1
    print("There are",ct,"number of items in",i,"category.")
    category[i]=ct
```

**USING IMPUTERS IN THIS DATA SET ON DIFFERENT DATA SET.**

IMPUTER ON NUMERICAL DATA.

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
df=data
```

```python
df
```

```python
# for imputer you have to use 2-d arrays.

df.iloc[:,1:3].values
```

IMPUTER ON CATEGORICAL DATA.

FOR CATEGORICAL WE USE MOST FREQUENT IN STRATEGY SECTION OF SimpleImputer.

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df.iloc[:,:1].values)
x=imputer.transform(df.iloc[:,:1].values)
```

```python
df.iloc[:,:1]=x
```

```python
df
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df.iloc[:,3:4].values)
x=imputer.transform(df.iloc[:,3:4].values)
```
